```python
import pandas as pd
```

```python
PATH = 'banknote_net.feather'
df = pd.read_feather(PATH)
df
```

```python
currency = df['Currency'].unique()
print(currency)
print(f'Length = {currency.__len__()}')
```

```python
denomination = df['Denomination'].unique()
print(denomination)
print(f'Length = {denomination.__len__()}')
```

```python
values = df.drop(columns=['Currency', 'Denomination'])
values.min()
```

```python
values.max()
```

```python
min = values.min().min()
max = values.max().max()
print(f'Most minimum = {min}')
print(f'Most maximum = {max}')
```

```python
for index, mean in values.mean().items():
    print(f'{index} = {mean}')
```

```python
for index, median in values.median().items():
    print(f'{index} = {median}')
```

```python
mean = values.mean().mean()
median = values.median().median()
print(f'Mean of means = {mean}')
print(f'Median of medians = {median}')
```
